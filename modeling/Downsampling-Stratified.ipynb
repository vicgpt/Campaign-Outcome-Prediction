```python
import pandas as pd
from sklearn.utils import resample
```

Due to high runtime of mixed effect models, we will reduce the training size to overcome this. Since we have unbalanced classes we will try to reduce dataset by reducing majority classes to improve upon the training

```python
train = pd.read_csv('../data/train_transformed.csv')
train.head()
```

```python
train.shape
```

```python
train_0 = train.query('term_deposit_subscribed == 0')
train_1 = train.query('term_deposit_subscribed == 1')
train_0.shape, train_1.shape
```

```python
train_0_ds = resample(train_0, replace=False, n_samples=len(train_1), random_state=100)
train_ds = pd.concat([train_0_ds, train_1])
train_ds.reset_index(drop=True, inplace=True)
train_ds
```

```python
train_ds.to_csv('../data/train_transformed_ds.csv')
```
